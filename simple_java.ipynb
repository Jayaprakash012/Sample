```python
public class SimplePrograms {

    // 1. Hello World
    static void helloWorld() {
        System.out.println("Hello, World!");
    }

    // 2. Add Two Numbers
    static void addNumbers() {
        int a = 10, b = 20;
        System.out.println("Sum = " + (a + b));
    }

    // 3. Even or Odd
    static void evenOdd() {
        int num = 7;
        if (num % 2 == 0)
            System.out.println("Even");
        else
            System.out.println("Odd");
    }

    // 4. Largest of Two Numbers
    static void largest() {
        int a = 15, b = 25;
        System.out.println("Largest = " + (a > b ? a : b));
    }

    // 5. Swap Two Numbers
    static void swap() {
        int a = 5, b = 10;
        int temp = a;
        a = b;
        b = temp;
        System.out.println("After swap: a = " + a + ", b = " + b);
    }

    // 6. Print 1 to 10
    static void printNumbers() {
        for (int i = 1; i <= 10; i++) {
            System.out.print(i + " ");
        }
        System.out.println();
    }

    // 7. Factorial
    static void factorial() {
        int num = 5, fact = 1;
        for (int i = 1; i <= num; i++) {
            fact *= i;
        }
        System.out.println("Factorial = " + fact);
    }

    // 8. Prime Number
    static void prime() {
        int num = 7;
        boolean isPrime = true;

        for (int i = 2; i <= num / 2; i++) {
            if (num % i == 0) {
                isPrime = false;
                break;
            }
        }

        System.out.println(isPrime ? "Prime" : "Not Prime");
    }

    // 9. Reverse Number
    static void reverse() {
        int num = 123, rev = 0;

        while (num != 0) {
            int digit = num % 10;
            rev = rev * 10 + digit;
            num /= 10;
        }

        System.out.println("Reversed = " + rev);
    }

    // 10. Fibonacci Series
    static void fibonacci() {
        int n = 10, a = 0, b = 1;

        System.out.print("Fibonacci: " + a + " " + b + " ");
        for (int i = 2; i < n; i++) {
            int c = a + b;
            System.out.print(c + " ");
            a = b;
            b = c;
        }
        System.out.println();
    }

    // Main Method
    public static void main(String[] args) {
        helloWorld();
        addNumbers();
        evenOdd();
        largest();
        swap();
        printNumbers();
        factorial();
        prime();
        reverse();
        fibonacci();
    }
}
```
